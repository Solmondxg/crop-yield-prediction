## PART 9: MODEL EVALUATION AND METRICS

### Evaluation Metrics Explained

**Mean Absolute Error (MAE)**
- Average of absolute differences between actual and predicted values
- Unit: Same as target variable (tons/hectare)
- Interpretation: "On average, predictions are off by X tons/hectare"
- Lower is better

**Mean Squared Error (MSE)**
- Average of squared differences
- Gives more weight to large errors
- Unit: Squared (tons/hectare)²
- Lower is better

**R² Score (Coefficient of Determination)**
- Percentage of variance explained by the model
- Range: 0 to 1 (or negative for poor models)
- Interpretation: "Model explains X% of the variation in yield"
- Closer to 1 is better (1.0 = perfect prediction)

2. **Handles interactions**: Understands how features work together

4. **Robust**: Less affected by outliers
5. **Higher accuracy**: Usually better than Linear Regression

**Hyperparameters to Tune:**
- `n_estimators`: Number of trees (default: 100)
- `max_depth`: Maximum depth of trees
- `min_samples_split`: Minimum samples to split a node

````python
# STEP 1: BUILD LINEAR REGRESSION MODEL
print("="*60)
print("MODEL 1: LINEAR REGRESSION")
print("="*60)

# Create and train the model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

print("\n✓ Model training completed!")
print(f"Model type: {type(lr_model).__name__}")

# STEP 2: VIEW MODEL COEFFICIENTS (FEATURE IMPORTANCE)
print("\n" + "-"*60)
print("FEATURE COEFFICIENTS (How much each feature affects yield)")
print("-"*60)

coefficients = pd.DataFrame({
    'Feature': feature_cols,
    'Coefficient': lr_model.coef_
}).sort_values('Coefficient', key=abs, ascending=False)

print(coefficients.to_string(index=False))

print(f"\nIntercept: {lr_model.intercept_:.4f}")
print("\nInterpretation:")
print("  - Positive coefficient: Feature increases → Yield increases")
print("  - Negative coefficient: Feature increases → Yield decreases")
print("  - Larger absolute value: Stronger effect on yield")

# STEP 3: MAKE PREDICTIONS
print("\n" + "-"*60)
print("MAKING PREDICTIONS")
print("-"*60)

# Predictions on training set
y_pred_train = lr_model.predict(X_train)

# Predictions on testing set
y_pred_test = lr_model.predict(X_test)

print(f"\nSample predictions on test set:")
print(f"{'Actual':>12} {'Predicted':>12} {'Difference':>12}")
print("-"*40)
for i in range(min(5, len(y_test))):
    diff = y_test.iloc[i] - y_pred_test[i]
    print(f"{y_test.iloc[i]:>12.2f} {y_pred_test[i]:>12.2f} {diff:>12.2f}")
````

**Simple Concept:**

**Equation:** `Yield = a₀ + a₁×Rainfall + a₂×Temperature + ... + aₙ×Feature_n`

**Limitations:**
